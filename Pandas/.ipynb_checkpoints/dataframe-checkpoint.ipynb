```python
df['half age']=df['age'].apply(lambda x : x/2)
df
```

```python
gpa_map={6.0:6,5.0:5,5.5:5}
df['gpa']=df['gpa'].map(gpa_map)
df
```

```python
#reshaping data 
```

```python
melttable=df.melt(id_vars=['employee_id','name'],
                 value_vars=['age','salary'],
                 var_name='metrics',
                 value_name='value')
melttable
```

```python
pivottable=melttable.pivot(index=['employee_id','name'],
                   columns=['metrics'],
                   values='value')
pivottable
```
